```python
import matplotlib.pyplot as plt
import numpy as np

# graph 1
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([10,5,8,9,4,12,12,7])
plt.plot(x,y)

x1=np.array([1,2,3,4,5,6,7,8])
y1=np.array([5,10,9,3,4,12,11,5])
plt.plot(x1,y1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# graph 1
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([10,5,8,9,4,12,12,7])
plt.scatter(x,y)

x1=np.array([1,2,3,4,5,6,7,8])
y1=np.array([5,10,9,3,4,12,11,5])
plt.scatter(x1,y1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
colors=np.array([10,20,30,40,50,60,70,80])
# graph 1
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([10,5,8,9,4,12,12,7])
plt.scatter(x,y,c=colors,cmap='viridis')

x1=np.array([1,2,3,4,5,6,7,8])
y1=np.array([5,10,9,3,4,12,11,5])
plt.scatter(x1,y1,c=colors,cmap='viridis')

plt.colorbar()
plt.show()
```
