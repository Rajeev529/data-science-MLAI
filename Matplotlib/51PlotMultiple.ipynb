```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([0,1,2,3,4])
x=np.array([0,1,2,3,4])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(1,2,1)
plt.plot(x,y)

y1=np.array([0,10,2,5,4])
x1=np.array([0,1,2,3,4])

plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.show()
```
