```python
# simple plot
import matplotlib.pyplot as plt
import numpy as np
data=np.array([60,10,25,4,1])
plt.pie(data)
plt.show()
```

```python
# labeling
import matplotlib.pyplot as plt
import numpy as np
data=np.array([60,10,25,4,1])
my_labels=['Nitrogen','oxygen','Argon','Other','Oxygen']
plt.pie(data,labels=my_labels)
plt.show()
```

```python
# StartAngle
import matplotlib.pyplot as plt
import numpy as np
data=np.array([60,10,25,4,1])
my_labels=['Nitrogen','oxygen','Argon','Other','Oxygen']
plt.pie(data, labels=my_labels, startangle=180)
plt.show()
```

```python
# explode
import matplotlib.pyplot as plt
import numpy as np
data=np.array([60,10,25,4,1])
my_labels=['Nitrogen','oxygen','Argon','Other','Oxygen']
my_explodes=[0.2,0.1,0,0,0]
# plt.pie(data, labels=my_labels, explode=my_explodes)
plt.pie(data, labels=my_labels, explode=my_explodes, shadow=True)
plt.show()
```

```python
# colors
import matplotlib.pyplot as plt
import numpy as np
data=np.array([60,10,25,4,1])
my_labels=['Nitrogen','oxygen','Argon','Other','Oxygen']
my_explodes=[0.2,0.1,0,0,0]
Mycolors=['hotpink','cyan','red','green', 'yellow']

plt.pie(data, labels=my_labels, explode=my_explodes, colors=Mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apple->35", "Banana->25", "Cherry->25", "mango->15"]

plt.pie(y, labels=mylabels)
plt.legend(title = "Fruits")
plt.show()
```
