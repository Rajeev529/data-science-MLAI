```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5,6])
y=np.array([14,4,19,8,9,0])
plt.plot(x,y,'o-.y',ms=10,mec='b',mfc='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
style={'family':'Times New Roman','color':'grey','size':20};
style1={'family':'Times New Roman','color':'black','size':10};

x=np.array([1,2,3,4,5,6])
y=np.array([180,200,250,220,270,240])

plt.plot(x,y,'o:y' ,ms=10,mec='r',mfc='b')

plt.xlabel('Days',fontdict=style)
plt.ylabel('AQI(delhi)',fontdict=style)

plt.title("Air Pollution Delhi",fontdict=style1, loc='right')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([1,2,3,4,5])
y1=np.array([10,9,12,11,19])

plt.plot(x1,y1)

x2=x1;
y2=np.array([11,12,9,1,12])
plt.plot(x2,y2,'o:y')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5]);
y=np.array([15,12,13,19,14])

plt.plot(x,y);
y1=np.array([15,10,20,17,11])
                                        #alpha=int used for transpacy
plt.bar(x,y1, alpha=.7, width=.7)

plt.bar(x,y, alpha=.7, width=.2)

plt.xlabel("x-Axis")
plt.ylabel("y-Axis")

plt.show()
```

```python
# here we going to draw pi charts 1.) data array, 2.) labels, 3.) explode, 4.) colors

import matplotlib.pyplot as plt
import numpy as np

data=np.array([30,20,10,10,30]);
labels=['china','united state','russia','ukrain', 'india']
explode=[.2, .1, 0.05, 0.05, 0.05]
colors=['red','green','purple','blue','pink']

plt.pie(data, labels=labels, explode=explode, colors=colors, shadow=True)
plt.show()
```
