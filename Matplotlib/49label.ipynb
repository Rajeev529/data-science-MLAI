```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([2,4,5,1,8])
plt.plot(x)
plt.title('Polution in delhi')
plt.xlabel('Days')
plt.ylabel('Polution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#fontdict
style1 = {'family':'serif', 'color':'blue', 'size':20}
style2={'family':'serif','color':'red','size':30}
x=np.array([2,4,5,1,8])
plt.plot(x,'o-b')
plt.title('Polution in delhi',fontdict=style2,loc='right')
plt.xlabel('Days',fontdict=style1)
plt.ylabel('Polution',fontdict=style1)
plt.show()
```
