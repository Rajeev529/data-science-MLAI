```python
import statistics as stat
import matplotlib.pyplot as plt

scores=[8,9,7,8,8,8,7,7]
idx=[0,1,2,3,4,5,6,7]

print(stat.stdev(scores)) # as the deviation is low as the data dicrete less away
scores=[9,6,7,1,8,8,5,5]

plt.scatter(scores,idx)
plt.show()
print(stat.stdev(scores))
```
