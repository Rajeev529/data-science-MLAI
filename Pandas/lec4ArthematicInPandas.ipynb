# Arthematics operation in pandas

```python
import pandas as pd
```

```python
dic={"before":[2,4,3,5,6],"after":[3,5,9,1,4]}
var=pd.DataFrame(dic)
var
```

```python
#addition by creating a new column in table
var["sum"]=var["before"]+var["after"]
var
```

```python
#substraction by creating a new column in table
var["substraction"]=var["before"]-var["after"]
var
```

```python
#divide by creating a new column in table
var["divide"]=var["before"]/var["after"]
var
```

```python
#comparision by creating a new column in table
var["istrue"]=var["substraction"]>=var["after"]
var
```
