# Create an Intervals index

```python
import pandas as pd
data={'name':['rajeev','sonu','monu','rahul'],'Time':["per hr","per hr","per hr","per hr"]}
idx=pd.IntervalIndex.from_breaks([1,1.5,3,4.5,6])
df=pd.DataFrame(data,index=idx)
df
```

```python
import pandas as pd
data={'name':['rajeev','sonu','monu','rahul'],'Time':["per hr","per hr","per hr","per hr"]}
idx=pd.IntervalIndex.from_tuples([(1,2),(2,3),(3,4),(4,5)])
df=pd.DataFrame(data,index=idx)
df
```

# Pandas: Indexing by date and time

```python
import pandas as pd
data={'name':['rajeev','sonu','monu','rahul'],'Time':["per hr","per hr","per hr","per hr"]}
idx=pd.date_range(start="2002-05-12 01:01:01", freq='h',periods=4)
df=pd.DataFrame(data,index=idx)
df
```
