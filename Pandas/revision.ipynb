```python
import pandas as pd
name=pd.Series(['akaash','rohit','vansh','mohit'])
marks=pd.Series([30,20,24,30])
internal=pd.Series([10,9,9,10])
df=pd.DataFrame({'name':name,'marks':marks,'internal':internal})
print(df)

sum=0
for i in range(0,4):
    sum+=df.loc[i,'marks']+df.loc[i,'internal']
    
print("average: ",sum/4)

print('number of row: ',len(df))
print('',df.size)
```

# dropna()

```python
import pandas as pd
import numpy as np
data = {
    'name': ['Alice', 'Bob', np.nan, 'David'],
    'score': [90, np.nan, 85, np.nan],
    'age': [25, 30, 22, np.nan]
}
df=pd.DataFrame(data)
```

```python
df['grade']=['A','A+','B+','C']
print("drop along hor: ")
print(df.dropna()) #drop each row have single nan values in it
print("drop along ver: ")
print(df.dropna(axis=1)) # drop each coln if single nan value apear atleast one time
```

```python
df.dropna(how='all') # how=all delete all the junck row if there is all nan value
```

```python
df.dropna(thresh=3,axis=0) # unko drop krdo jinme bhi 3 non nanvalues nhi hai horizontaly
```

```python
df.dropna(thresh=3,axis=1) # unko drop krdo jinme bhi 3 non nanvalues nhi hai vertically
```

# fillna()

```python
import pandas as pd
import numpy as np
data = {
    'name': ['Alice', 'Bob', np.nan, 'David'],
    'score': [90, np.nan, 85, np.nan],
    'age': [25, 30, 22, np.nan]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.fillna(-1) #fill entire nan values from the -1
```

```python
# Fill NaNs in score column with the mean:
df['score']=df['score'].fillna(df['score'].mean())
df
```

```python
df.ffill()
```

```python
df.bfill()
```

```python
df.ffill(axis=1) # age ka last coln fill hogya hai score k last coln se
```
