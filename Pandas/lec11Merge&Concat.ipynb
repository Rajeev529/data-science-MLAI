# merging 2 data frame

```python
import pandas as pd
var1=pd.DataFrame({'Even':[2,4,6,8],'Natural':[1,2,3,4]})
var2=pd.DataFrame({'Even':[2,4,6,8],'Odd':[1,3,5,7]})
pd.merge(var1,var2,on="Even") #0n is used for taking common column in both
```

```python
var3=pd.DataFrame({'A':[2,4,6,8],'B':[1,2,3,4]})
var4=pd.DataFrame({'A':[2,4,6,10],'C':[1,3,5,7]})
pd.merge(var3,var4,how="right") #how="right" or "left" or "outer"
```

```python
pd.merge(var3,var4,how="left") 
```

```python
pd.merge(var3,var4,how="outer",indicator=True) 
```

# concat

```python
val4=pd.DataFrame({'A':[1,2,3,4],'B':[2,4,6,8]})
val5=pd.DataFrame({'A':[5,6,7,8],'B':[10,12,14,16]})
pd.concat([val4,val5])
```

```python
pd.concat([val4,val5],axis=1)
```

```python
pd.concat([val4,val5],keys=['v4','v5'])
```

# join

```python
val4=pd.DataFrame({'A':[1,2,3,4],'B':[2,4,6,8]})
val5=pd.DataFrame({'D':[5,6,7,8],'C':[10,12,14,16]})
val4.join(val5) # All the column head must be diffrent else it will show error
```

```python
val4=pd.DataFrame({'A':[1,2,3,4],'B':[2,4,6,8]},index=['a','b','c','d'])
val5=pd.DataFrame({'D':[5,6],'C':[10,12]},index=['a','b'])
val5.join(val4) # All the column head must be diffrent else it will show error
```

```python
# to use two same column you have to pass lsuffix('name')
val4=pd.DataFrame({'A':[1,2,3,4],'B':[2,4,6,8]},index=['a','b','c','d'])
val5=pd.DataFrame({'D':[5,6],'C':[10,12]},index=['a','b'])
val4.join(val5) 
```

```python
# to use two same column you have to pass lsuffix='name'
val4=pd.DataFrame({'A':[1,2,3,4],'B':[2,4,6,8]},index=['a','b','c','d'])
val5=pd.DataFrame({'D':[5,6],'B':[10,12]},index=['a','b'])
val4.join(val5,lsuffix='name') #pass how="left","right","inner","outer"
```
