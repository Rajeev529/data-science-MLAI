```python
import pandas as pd
file=pd.read_csv(r"C:\Users\hp\Data Science ML\webscraping\datacollected.csv")
arredx=[]
arrlv=[]
arrcs=[]
arrsl=[]
for i in skills:
    data=file[file['course']==i]
    # print(data)
    edx=data[data['platform']=='edx']
    # print(edx)
    lv=data[data['platform']=='learnvern']
    cs=data[data['platform']=='coursera']
    sl=data[data['platform']=='simplilearn']
    for indx, row in edx.iterrows():
        arredx.append(row)
    for indx, row in lv.iterrows():
        arrlv.append(row)
    for indx, row in cs.iterrows():
        arrcs.append(row)
    for indx, row in sl.iterrows():
        arrsl.append(row)
    
```

```python

final=[]
for i in range(0, len(arrcs)):
    if(i>=0 and i<len(arrcs)):
        final.append(arrcs[i])

    if(i>=0 and i<len(arrsl)):
        final.append(arrsl[i])
        
    if(i>=0 and i<len(arredx)):
        final.append(arredx[i])
        
    if(i>=0 and i<len(arrlv)):
        final.append(arrlv[i])
```
