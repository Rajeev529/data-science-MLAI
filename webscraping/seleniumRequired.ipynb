```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

service = Service(executable_path=r"C:\Users\hp\Data Science ML\webscraping\chromedriver.exe")
driver = webdriver.Chrome(service=service)

driver.get("https://www.amazon.in/")

time.sleep(3)

search_box = driver.find_element(By.ID, "twotabsearchtextbox")
search_box.send_keys("ps5 games")
driver.find_element(By.ID, "nav-search-submit-button").click()

time.sleep(5)  # Increased sleep to ensure search results load

# Find all product cards on the page.
# Amazon uses 'div[data-component-type="s-search-result"]' as a reliable selector for individual product listings.
product_cards = driver.find_elements(By.CSS_SELECTOR, 'div[data-component-type="s-search-result"]')

print(f"Found {len(product_cards)} product cards.")

for card in product_cards:
    title = 'N/A'
    reviews = 'N/A'
    price = 'N/A'

    try:
        # Title is usually within a span with class 'a-size-medium a-color-base a-text-normal' inside h2.
        title_element = card.find_element(By.CSS_SELECTOR, 'h2 a.a-link-normal span.a-size-medium.a-color-base.a-text-normal')
        title = title_element.text
    except Exception:
        pass

    try:
        # Review count is usually in a span with class 'a-size-base s-underline-text'
        reviews_element = card.find_element(By.CSS_SELECTOR, 'span.a-size-base.s-underline-text')
        reviews = reviews_element.text
    except Exception:
        pass

    try:
        # Price is often found in 'span.a-price' containing 'span.a-offscreen' or 'span.a-price-whole'
        price_element_offscreen = card.find_element(By.CSS_SELECTOR, 'span.a-price span.a-offscreen')
        price = price_element_offscreen.get_attribute('textContent')
    except Exception:
        try:
            # Fallback for price if 'a-offscreen' is not found (e.g., when price is split)
            price_whole = card.find_element(By.CSS_SELECTOR, 'span.a-price-whole').text
            price_fraction = card.find_element(By.CSS_SELECTOR, 'span.a-price-fraction').text
            price = f"₹{price_whole}{price_fraction}"
        except Exception:
            pass
            
    print(f"Title: {title}")
    print(f"Reviews: {reviews}")
    print(f"Price: {price}")
    print("-" * 30)

driver.quit()
```
