```python
import pandas as pd

# Read the CSV file
file = pd.read_csv(r"C:\Users\hp\Data Science ML\webscraping\datacollected.csv", index_col=False)

# User input
user = input("Search here: ")

# Create a list to store filtered DataFrames
arr = []

# Check if 'head' column exists and search it
if 'head' in file.columns:
    arr.append(file[file['head'].str.contains(user, case=False, na=False)])

# Check if 'description' column exists and search it
if 'description' in file.columns:
    arr.append(file[file['description'].str.contains(user, case=False, na=False)])

# Concatenate and drop duplicates
if arr:
    df = pd.concat(arr).drop_duplicates().reset_index(drop=True)
    print(df)
else:
    print("No matching results found.")
```
