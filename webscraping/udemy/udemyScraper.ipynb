```python
# udemy scraping explore

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

service = Service(executable_path=r"C:\Users\hp\Data Science ML\webscraping\chromedriver.exe")
driver = webdriver.Chrome(service=service)

driver.get("https://www.udemy.com/courses/marketing")
time.sleep(5)  # Let the page load

cards = driver.find_elements(By.XPATH, '//div[@class="course-list_card-layout-container__F2SfZ"]')

script=[]

des=[]
heading=[]
link=[]
revw=[]
price=[]
img=[]

for i in cards:
    
    des1 = i.find_element(By.XPATH, './/p[@class="ud-text-sm course-card_course-headline__BBr5m"]').text.strip()
    des.append(des1)
    
    heading1=i.find_element(By.XPATH,'.//h3[@ data-purpose="course-title-url"]//a')
    heading.append(heading1.text.strip())
    
    link1=heading1.get_attribute('href')
    link.append(link1)
    
    revw1=i.find_element(By.XPATH,'.//span[@ class="ud-text-xs course-card-ratings_reviews-text__rx1CN"]').text.strip()
    revw.append(revw1)
    
    price1 = i.find_element(By.XPATH, './/div[contains(@class, "price-text-container")]//span/span')
    price.append(price1.text.strip())

    images=i.find_element(By.XPATH,'.//img')
    img1=images.get_attribute('src')
    img.append(img1)
driver.quit()
```

```python
# udemy cleaning data
newprice=[]
for i in range (0,len(price)):
    a=price[i].replace('₹',"")
    b=a.replace(',',"")
    newprice.append(int(b))

review=[]
for i in range(0,len(revw)):
    a=revw[i].replace('(',"")
    b=a.replace(',',"")
    c=b.replace(')',"")
    review.append(int(c))

import pandas as pd
data={"Heading":heading, "Description": des, "Revewe":review, "URL": link, "price": newprice, "image":img}
# pd.DataFrame(data).to_csv('digitalMarketingP1.csv')
pd.DataFrame(data)
```
