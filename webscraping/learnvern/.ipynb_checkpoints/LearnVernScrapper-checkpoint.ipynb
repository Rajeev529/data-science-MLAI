```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

import numpy as np

service=Service(executable_path=r"C:\Users\hp\Data Science ML\webscraping\chromedriver.exe")
driver=webdriver.Chrome(service=service)

driver.get("https://www.learnvern.com/courses?txt_search=marketing")
time.sleep(10)

cards=driver.find_elements(By.XPATH,'//div[@class="course-item d-inline-block align-items-stretch"]')

head=[]
link=[]
revw=[]
img=[]
price=[]
lang=[]

for i in cards:
    heading=i.find_element(By.XPATH,'.//a[@class="card-title courses-card-title"]')
    head.append(heading.text.strip())
    url=link.append(heading.get_attribute('href'))

    xpath1='.//div[@class="d-flex flex-row cat-course-meta align-items-center mt-auto"]//span'
    revw1=i.find_element(By.XPATH,xpath1).text.strip()
    revw.append(revw1)

    img1=i.find_element(By.XPATH,'.//img[@class="card-img-top d-flex mx-auto"]')
    url1=img1.get_attribute('src')
    img.append(url1)
    try:
        price1=i.find_element(By.XPATH,'.//label[@class="free-badge"]')
        price.append(price1.text.strip())
    except:
        price.append(np.nan)

driver.quit()
```

# data cleaning

```python
import numpy as np
review=[]
for i in revw:
    review.append(int(i))

newprice=[]
for i in price:
    if(i=='FREE'):
        newprice.append(np.nan)
    elif not np.isnan(i) :
        newprice.append(int(i))
    else:
        newprice.append(np.nan)
newprice

import pandas as pd
data={"Heading":head, "Revewe":review, "URL": link, "price": newprice, "image":img}
file=pd.DataFrame(data)
file
```

```python
pd.DataFrame(data).to_csv("marketingLVp1.csv")
```
