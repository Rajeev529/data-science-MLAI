```python
import pandas as pd
import matplotlib.pyplot as plt
# data=pd.read_csv(r"C:\Users\hp\Downloads\Student_Performance.csv")
data=pd.read_csv(r"C:\Users\hp\Downloads\Student_Performance.csv",nrows=200)
data
```

```python
sample_qpaper=data['Sample Question Papers Practiced']
score=data['Performance Index']
plt.xlabel("sample paper")
plt.ylabel("performance")
plt.scatter(sample_qpaper, score, color='black')
```

```python
def grad_dec(m, c, data, l1):
    dm = 0
    dc = 0
    n = len(data)

    for i in range(n):
        x = data.iloc[i]['Sample Question Papers Practiced']
        y = data.iloc[i]['Performance Index']
        dm += (-2/n) * x * (y - (m * x + c))
        dc += (-2/n) * (y - (m * x + c))

    # Update parameters using learning rate l1
    m = m - l1 * dm
    c = c - l1 * dc

    return m, c
```

```python
def loss_function(m, c, data):
    error = 0
    n = len(data)
    for i in range(n):
        x = data.iloc[i]['Sample Question Papers Practiced']
        y = data.iloc[i]['Performance Index']
        error += (y - (m * x + c)) ** 2
    return error / n
```

```python
import matplotlib.pyplot as plt

# Initial parameters
m = 0
c = 0
l = 0.001
epochs = 1000

# Gradient descent loop
for i in range(epochs):
    if i % 50 == 0:
        print(f"Epoch {i}: Loss = {loss_function(m, c, data):.4f}")
    m, c = grad_dec(m, c, data, l)

# Plotting
sample_qpaper = data['Sample Question Papers Practiced']
score = data['Performance Index']

plt.xlabel("Sample Question Papers Practiced")
plt.ylabel("Performance Index")
plt.scatter(sample_qpaper, score, color='black')

x_axis = list(range(0, 15))
y_axis = [m * x + c for x in x_axis]

plt.plot(x_axis, y_axis, color='red')
plt.show()
```
