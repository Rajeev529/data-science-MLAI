```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r'C:\Users\hp\Downloads\dataset for ml\play_tennis.csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
outlook=LabelEncoder()
temp=LabelEncoder()
humidity=LabelEncoder()
play=LabelEncoder()
```

```python
data['outlook']=outlook.fit_transform(data['outlook'])
data['temp']=outlook.fit_transform(data['temp'])
data['humidity']=outlook.fit_transform(data['humidity'])
data['play']=outlook.fit_transform(data['play'])
```

```python
data
```

```python
feature_col=['outlook', 'temp', 'humidity']
x=data[feature_col]
y=data['play']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2)
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='gini')
classifier.fit(x_train, y_train)
```

```python
classifier.predict(x_train)
```
