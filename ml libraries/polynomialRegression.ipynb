```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv(r'C:\Users\hp\Downloads\potato_price.csv')
data
```

```python
year=[]
n=len(data)
for i in range(int(n)):
    s=""
    date=data.iloc[i]['Date']
    # print(rev)
    for j in date:
        if j=='/':
            continue
        else:
            s=s+j
    print(s)
    year.append(int(s))
```

```python
data['Date']=year
data
```
