# sorting in array

```python
# ✅ 1. np.sort(): Returns a sorted copy of the array
```

```python
import numpy as np
arr=np.array([8,5,7,6,4,1,2])
arrCopy=np.sort(arr)
print(arrCopy)
```

```python
import numpy as np
arr2d = np.array([[4, 1], [3, 2]])
arrCopy=np.sort(arr2d, axis=1)  # horizontally sort
print(arrCopy)
arrCopy=np.sort(arr2d,axis=0)   # vertically sort
print(arrCopy)
```

```python
# ✅ 3. np.argsort(): Returns the indices that would sort the array
```

```python
import numpy as np
arr=np.array([10,40,30])

idxs=np.argsort(arr)
print(idxs)

```

```python
# 4. np.lexsort((arr1,arr2)): Sort by multiple keys
```

```python
import numpy as np
age=np.array([10,40,30])
names=np.array(['luke','marie','bob'])
idxs=np.lexsort((age,names))
print(idxs)
print(names[idxs])
```

```python
import pandas as pd
import numpy as np

# Step 1: Read the CSV file
df = pd.read_csv(r"C:\Users\hp\Data Science ML\Pandas\udated_marksheet.csv")  # replace with your actual filename

# Step 2: Convert the 'total_marks' column to a NumPy array
marks = df['total_marks'].to_numpy()

# Step 3: Use lexsort to get sorted indices (ascending order)
# Note: lexsort sorts using the LAST key as the primary sort key
sorted_indices = np.lexsort((df['name'].to_numpy(), marks))

# Step 4: Reorder the DataFrame using sorted indices
sorted_df = df.iloc[sorted_indices]

# Step 5: Print or save the result
print(sorted_df)
# sorted_df.to_csv('sorted_output.csv', index=False)
```
